```python
import pandas as pd
pd.set_option('display.max_rows', None)
```

```python
df = pd.read_csv('Bollywood.csv')
```

```python
df.columns
```

```python
#Change column name Lead Star to Lead_Star Movie Name to Movie_Name
df.rename(columns={'Lead Star':'Lead_Star',
                    'Movie Name':'Movie', 
                    'Release Period': 'Release_Period',
                    'Whether Remake': 'Whether_Remake',
                    'Whether Franchise': 'Whether_Franchise',
                    'New Actor': 'New_Actor',
                    'New Director': 'New_Director',
                    'New Music Director': 'New_Music_Director',
                    'Music Director':'Music_Director',
                    'Number of Screens': 'Screens',
                    'Budget(INR)': 'Budget',
                    'Revenue(INR)': 'Revenue',
                    'ProfitFactor': 'ProfitFactor',
                    'Rating': 'IMDb'},inplace=True)
```

```python
#change Genre love_story to Romance and rom_com to Romance
df['Genre'] = df['Genre'].replace(['love_story', 'rom__com'], 'romance')



#change all Genre to Sentence case
df['Genre'] = df['Genre'].str.title()
```

```python
#sort df by Rating
df = df.sort_values(by='IMDb', ascending=False)

#remove columns
df.drop(['Release_Period', 'Whether_Remake', 'Whether_Franchise', 'New_Director', 'New_Music_Director', 'Music_Director', 'ProfitFactor'], axis=1, inplace=True)

#rename columns
df.rename(columns={'Movie_Name':'Movie', 'Lead_Star': 'Star', 'New_Actor': 'Newbie'}, inplace=True)
```

```python
df = df[df['Year']>2000]
```

```python
df = df.sort_values(by='Year', ascending=True)
df = df[df['Star']=='Shahrukh Khan']
```
